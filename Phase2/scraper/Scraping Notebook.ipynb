```python
restoreIP = ["https://leetcode.com/problems/restore-ip-addresses/discuss/?currentPage=1&orderBy=hot&query=&tag=c-3"
            ]
```

```python
import selenium
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium import webdriver
```

```python
c = 1
```

```python
from selenium.webdriver import FirefoxOptions
import re
# from selenium.webdriver.support import expected_conditions as EC
for link in restoreIP:

    opts = FirefoxOptions()
    opts.add_argument("--headless")
    # browser = webdriver.Firefox(firefox_options=opts)


    wd = webdriver.Firefox(firefox_options=opts, executable_path = './geckodriver')
    wd.get(link)

    # Wait for the dynamically loaded elements to show up
    WebDriverWait(wd, 100).until(
        EC.visibility_of_element_located((By.CLASS_NAME, "topic-title__3LYM")))


    # And grab the page HTML source
    html_page = wd.page_source
    wd.quit()

    # Now you can use html_page as you like
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html_page)
    
    links = soup.findAll(class_ = "title-link__1ay5")
    # print(links)
    for link in links:
        link_n = "https://leetcode.com" + link.get_attribute_list("href")[0]
        print(link_n)

        wd = webdriver.Firefox(firefox_options=opts, executable_path = './geckodriver')
        wd.get(link_n)

        WebDriverWait(wd, 10).until(
        EC.visibility_of_element_located((By.CLASS_NAME, "discuss-markdown-container")))

        html_page = wd.page_source
        wd.quit()

        soup = BeautifulSoup(html_page)
        
        codes = soup.findAll("code")
        for code in codes:
            with open("dataset/restoreIP_"+str(c)+"__NEW.c", "w", encoding="utf-8") as f: 
                # Writing data to a file 
                f.write(str(code.text)) 
                print(str(code.text))
                c += 1
```
